```python
import sounddevice as sd
import numpy as np #NumPyは基本必須です。インポートしてください。
import wave #オーディオファイルを扱いたい場合インポートします
fs = 192000
sd.default.samplerate = fs
sd.query_devices()
```

```python
sd.default.device = 4
# sd.default.device = [4, 5]として入出力をそれぞれ設定することもできます。
```

```python
#信号のリスト作成
import scipy.io as sio

pi = np.pi
f = 40000
filename ='/Users/shota/Documents/klo_lab/matlab/phase/20200824/w_-20phase-10.mat'
phase_m = sio.loadmat(filename)
phase_list = phase_m['phix']
sig = []
for phase in phase_list:
    sin = np.sin(2*pi*f*np.arange(0,int(fs))/fs+phase)
    sig.append(sin)
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for i,s in enumerate(sig):
    sigF = np.fft.fft(s, n=fs)

    fig = plt.figure(i)
 
    ax1 = fig.add_subplot(211,title = str(i))
    ax2 = fig.add_subplot(212)
    ax1.plot(20*np.log10(sigF[0:int(fs/2)]))
    ax2.plot(np.arange(0,int(fs/f)+2)/fs,s[0:int(fs/f)+2])
    plt.show()
```

```python
#アンプfullで20Vppになるように調整
sig_list = np.zeros((fs,28))
sig_list[:,0] = sig[0]*0.20
sig_list[:,1] = sig[1]*0.40
sig_list[:,2] = sig[2]*0.97
sig_list[:,3] = sig[3]*1.28
sig_list[:,14] = sig[4]*0.71
sig_list[:,15] = sig[5]*0.84
sig_list[:,16] = sig[6]*1.8
sig_list[:,17] = sig[7]*2.20
```
